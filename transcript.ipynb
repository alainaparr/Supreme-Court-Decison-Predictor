```python
# NEW CELL
# This cell checks to see if I'm properly referencing the files I want on the website I want
import requests
from bs4 import BeautifulSoup
import os
from urllib.parse import urljoin

base_url = "https://www.supremecourt.gov/oral_arguments/argument_transcript/2021"
reqs = requests.get(base_url)
soup = BeautifulSoup(reqs.text, "html.parser")

pdf_links = []
for link in soup.find_all("a"):
    href = link.get("href")
    if href and href.endswith(".pdf"):
        absolute_url = urljoin(base_url, href)
        pdf_links.append(absolute_url)
        filename = os.path.basename(absolute_url)
        file_path = os.path.join("/content/gdrive/MyDrive/supreme court/2021", filename)
        print(f"Downloading: {filename}")
        response = requests.get(absolute_url)
        with open(file_path, "wb") as file:
            file.write(response.content)

print("Downloads completed.")

```

```python
import os
import gdown
from docx import Document
import shutil

# Path to the folder containing the Google Docs files
folder_path = '/content/gdrive/MyDrive/supreme court/2020'

# New folder path to store the converted text files
new_folder_path = os.path.join(folder_path, '2020 text test')

# Create the new folder if it doesn't exist
if not os.path.exists(new_folder_path):
    os.makedirs(new_folder_path)

# Function to convert Google Docs to text
def convert_gdoc_to_text(file_id):
    gdoc_url = f'https://docs.google.com/document/d/{file_id}/export?format=docx'
    temp_file = '/tmp/gdoc.docx'  # Temporary file to store the downloaded Google Docs file
    try:
        gdown.download(gdoc_url, temp_file, quiet=False)
    except Exception as e:
        print(f"Error downloading Google Docs file: {file_id}")
        return ''

    doc = Document(temp_file)
    text = ''
    for paragraph in doc.paragraphs:
        text += paragraph.text + '\n'

    os.remove(temp_file)  # Delete the temporary file

    return text

# Iterate over files in the folder
for root, dirs, files in os.walk(folder_path):
    for file_name in files:
        print(file_name)
        if file_name.endswith('.gdoc'):
            file_path = os.path.join(root, file_name)

            # Get the file ID from the file name (assuming file name is the ID)
            file_id = os.path.splitext(file_name)[0]

            # Convert Google Docs to text
            text = convert_gdoc_to_text(file_id)
            if text:
                # Save the text to a file with the same name as the Google Docs
                # in the new folder with a .txt extension
                new_file_path = os.path.join(new_folder_path, file_name)
                new_file_path = os.path.splitext(new_file_path)[0] + '.txt'
                with open(new_file_path, 'w') as text_file:
                    text_file.write(text)
```

```python
import os
import gdown
import subprocess
import shutil
import PyPDF2

# Path to the folder containing the Google Docs files
folder_path = '/content/gdrive/MyDrive/supreme court/2020'

# New folder path to store the converted text files
new_folder_path = os.path.join(folder_path, '2020 text test')

# Create the new folder if it doesn't exist
if not os.path.exists(new_folder_path):
    os.makedirs(new_folder_path)

# Function to convert Google Docs to PDF
def convert_gdoc_to_pdf(file_id):
    gdoc_url = f'https://docs.google.com/document/d/{file_id}/export?format=pdf'
    temp_file = '/tmp/gdoc.pdf'  # Temporary file to store the downloaded Google Docs file
    try:
        gdown.download(gdoc_url, temp_file, quiet=False)
    except Exception as e:
        print(f"Error downloading Google Docs file: {file_id}")
        return ''

    return temp_file
 # Function to convert PDF to text using PyPDF2
def convert_pdf_to_text(file_path):
    try:
        with open(file_path, 'rb') as pdf_file:
            reader = PyPDF2.PdfFileReader(pdf_file)
            text = ''
            for page in reader.pages:
                text += page.extract_text()
            return text
    except Exception as e:
        print(f"Error converting PDF: {file_path}")
        return ''

# Iterate over files in the folder
for root, dirs, files in os.walk(folder_path):
    for file_name in files:
        if file_name.endswith('.gdoc'):
            print(file_name)
            file_path = os.path.join(root, file_name)

            # Get the file ID from the file name (assuming file name is the ID)
            file_id = os.path.splitext(file_name)[0]

            # Convert Google Docs to PDF
            pdf_file_path = convert_gdoc_to_pdf(file_id)
            if pdf_file_path:
                # Convert PDF to text
                text = convert_pdf_to_text(pdf_file_path)
                if text:
                    # Save the text to a file with the same name as the Google Docs
                    # in the new folder with a .txt extension
                    new_file_path = os.path.join(new_folder_path, file_name)
                    new_file_path = os.path.splitext(new_file_path)[0] + '.txt'
                    with open(new_file_path, 'w') as text_file:
                        text_file.write(text)
                # Remove the temporary PDF file
                os.remove(pdf_file_path)   
```

```python
import os
import shutil
from pdfminer import high_level

# Path to the folder containing the PDF files
folder_path = '/content/gdrive/MyDrive/supreme court/2020'

# New folder path to store the converted text files
new_folder_path = os.path.join(folder_path, '2020 text test')

# Create the new folder if it doesn't exist
if not os.path.exists(new_folder_path):
    os.makedirs(new_folder_path)

# Function to convert PDF to text using pdfminer.six
def convert_pdf_to_text(file_path):
    try:
        text = high_level.extract_text(file_path)
        return text
    except Exception as e:
        print(f"Error converting PDF: {file_path}")
        return ''

# Iterate over files in the folder
for root, dirs, files in os.walk(folder_path):
    for file_name in files:
        print(file_name)
        if file_name.endswith('.pdf'):
            file_path = os.path.join(root, file_name)

            # Convert PDF to text
            text = convert_pdf_to_text(file_path)
            if text:
                # Save the text to a file with the same name as the PDF
                # in the new folder with a .txt extension
                new_file_path = os.path.join(new_folder_path, file_name)
                new_file_path = os.path.splitext(new_file_path)[0] + '.txt'
                with open(new_file_path, 'w') as text_file:
                    text_file.write(text)
```
